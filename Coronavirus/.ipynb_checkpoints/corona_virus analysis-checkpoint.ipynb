#### Coronavirus Data 
Fin Mars et déjà des milliers de victimes à travers le monde. Je me propose en ce jour d'effectuer un aperçu sur le virus le plus populaire de l'année 2020. Nous récupérons les données récoltées par l'université Johns Hopkins et mises à la dsposition du public. 

```python
import pandas as pd 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# Importation des données
confirme = pd.read_csv('confirmed.csv')
morts = pd.read_csv('deaths.csv')
gueris = pd.read_csv('recovered.csv')
```

```python
# Aperçu des données  
confirme.tail()
```

Nous constatons que nous avons les dates en colonne. Il serait beaucoup plus pratique de les avoir en ligne afin de pouvoir facilement avoir un visuel evolution sur le temps. De plus pour le moment il ne nous apparait pas utile de maintenir les colonnes de repérages (longitude, latitude, Province/Etat) car nous n'avons aucun objectif de placer ces point sur une carte.

```python
confirme = confirme.groupby(confirme['Country/Region']).sum().drop(['Lat','Long'], axis=1).T
morts = morts.groupby(morts['Country/Region']).sum().drop(['Lat','Long'], axis=1).T
gueris = gueris.groupby(gueris['Country/Region']).sum().drop(['Lat','Long'], axis=1).T
```

```python
# Nouvel aperçu des données 
gueris.tail()
```

Maintenant nous nous proposons de savoir quel est le pays avec le plus grand nombre de cas actifs au jour d'aujourd'hui.

```python
actif = confirme.copy()
```

```python
for jour in range(1,len(confirme)):
    actif.iloc[jour] = confirme.iloc[jour] - morts.iloc[jour-1] - gueris.iloc[jour-1]
```

```python
actif.tail(n=1).T.sort_values(by='3/27/20', ascending = False).head(n=10)
```

A ce jour les pays ci-dessus sont les pays ayant le plus grand nombre de malades répertoriés.
Nous allons porté notre intérêt de savoir quel est le pays qui a l'évolution la plus rapide parmis ces pays les plus touchés. 

```python
pays = ['China','Italy','Iran','US','Spain','Germany','France','Switzerland','United Kingdom','Netherlands','Austria']
```

```python
actif[pays].plot()
```

On constate que depuis le debut du mois de Mars effectivement la Chine, n'est plus le pays ayant le plus grand nombre de cas actifs. On constate même qu'aujourd'hui, le pays ne figure pas parmis les 10 pays les plus touchés. Etant Camerounais, et juste pour le fun je me demande quel est l'aperçu de l'évolution des cas dans mon pays d'origine.

```python
actif['Cameroon'].plot()
plt.grid(True)
plt.show()
```

Pour finir voici un visuel sur les pays ayant le plus grand nombre de morts de la journée d'hier.

```python
mort_jour = morts.copy()
for jour in range(1,len(mort_jour)):
    mort_jour.iloc[jour] = morts.iloc[jour]  - morts.iloc[jour-1] 
```

```python
mort_jour.T['3/27/20'].sort_values(ascending = True).tail(n=15).plot(kind='bar')
plt.grid(True)
```
